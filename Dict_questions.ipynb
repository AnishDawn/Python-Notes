```python
#Solution
S = input("Enter your sentence: ")

S = ''.join([char for char in S if char.isalnum() or char.isspace()])

word_list = S.split()
print(word_list)
dict_count = {}
for i in word_list:
    dict_count[i] = word_list.count(i)

for i, j in dict_count.items():
    print('{}:{}'.format(i, j))
```

```python
"""

Print keys in alphabetic order

"""

Ni = input("Enter student names separated by comma: ")
Di = input("Enter the IDs separated by comma: ")

Ni_list = Ni.split(',')
Di_list = Di.split(',')

dict_students = {}

for i in range(len(Ni_list)):
    dict_students[Ni_list[i]] = Di_list[i]

print(dict_students)

stud = dict_students.items()
# print(stud)

stud_list = list(stud)
stud_list.sort()
# print(stud_list)

for i in stud_list:
    print("{} {}".format(i[0], i[1]))
```

```python
#Solution
def list_sum (n_list):
    length = len(n_list)
    if length == 1:
        return
    else:
        m_list = []
        for i in range(length - 1):
            m_list.append(n_list[i] + n_list[i+1])
        print(m_list)
    return list_sum(m_list)


num_string = input("Enter comma-separated integers: ")
num_list = num_string.split(",")
N_list = [int(i) for i in num_list]
print(N_list)

list_sum(N_list)


    
```
