```python
#Solution
def check_prime(N):
    count = 0
    for i in range(N):
        if N % (N - i) == 0:
            count += 1
    if count == 2:
        return "Prime Number"
    else:
        return "Not a Prime Number"

N = int(input("Enter a number: "))
print(check_prime(N))
```

```python
#Solution

S = input("Enter a word: ")

upper_count = 0
lower_count = 0
for i in S:
    if i.isupper():
        upper_count += 1
    else:
        lower_count += 1

print("Uppercase letters are ", upper_count)
print("lowercase letters are ", lower_count)
```

```python
#Solution

def fizz_buzz(num):
    if num % 5 == 0 and num % 3 == 0:
        return print("FizzBuzz!")
    elif num % 3 == 0:
        return print("Fizz!")
    elif num % 5 == 0:
        return print("Buzz!")
    else:
        return print(num)
num = int(input())
fizz_buzz(num)

```
